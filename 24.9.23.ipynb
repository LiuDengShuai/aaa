```python
import matplotlib.pyplot as plt
import numpy as np
x=np.linspace(0,20,100)#0到20之间生成一百个数
y=np.sin(x)
plt.scatter(x,y)
plt.axhline(y=0, ls='--',color='r')
plt.text(3.5,0,'sin(x)',weight='bold',c='r')
```

```python
x = np.random.normal(0,0.8,1000)
y = np.random.normal(-3,1,1000)
z = np.random.normal(3,2.,1000)
```

```python
kwargs = dict(alpha=0.3, bins=40,density=True)
```

```python
plt.hist(x, **kwargs)
plt.hist(y,**kwargs)
plt.hist(z,**kwargs)
```
